### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada gráfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades más pobladas, con los numeros de las poblaciones (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)
2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
3. El scatter de área de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# GRÁFICO DE BARRAS 1. Las ciudades más pobladas, con los numeros de las poblaciones 
# (tendrás que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrón, con trasparencia)

cities = dataset.sort_values(by="population_total", ascending=False).head(5)
cities
```

```python
plt.bar(cities["city"], cities["population_total"])
```

```python
plt.bar(cities["city"], cities["population_total"], color = ["blue","grey","orange","olive","brown"])
```

```python
#plt.figure(figsize=(8,8))
plt.xlabel("Ciudades")
plt.ylabel("Habitantes (miles)")
plt.title("Ciudades más pobladas de California")
#Çplt.ylim(0, 5000)  
#plt.yticks(np.arange(0, 5001,1000)) Me cambia toda la barra cuando cambio el rango


bars = plt.bar(cities["city"], cities["population_total"], color = ["blue","grey","orange","olive","brown"], alpha=0.5)


for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(
        x=bar.get_x() + bar.get_width() /2,
        y=height + 100000,
        s=height/1000,
        ha="center",
        va="bottom"
    )
```

```python
#2. Histograma del área total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamaño 18, color azul acero)
plt.hist(dataset["area_total_km2"], bins=50, alpha= 0.5, color="steelblue");
plt.xlabel("Superficie (km2)")
plt.ylabel("Número de ciudades")
plt.legend(loc="upper right", title="Hist area");
```

```python
#3. El scatter de área de agua de un ciudad con su poblacion.
plt.scatter(dataset["area_water_km2"], dataset["population_total"],c= "g", alpha=0.3)
plt.xlabel("Área de agua en la ciudad (km2)")
plt.ylabel("Población (miles habitantes)")
plt.title("Población por kilómetro cuadrado de agua")
```

```python
#4. El boxplot de la altura de las ciudades.
plt.boxplot(dataset["elevation_m"], vert=False, patch_artist=True,boxprops=dict(facecolor='red', color='black'))
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altidtud de las ciudades de California")
```

```python
#Todos juntos con subplots

fig, axs = plt.subplots(2, 2, figsize=(16, 16))
axs[0,0].bar(cities["city"], cities["population_total"], color=["blue","grey","orange","olive","brown"], alpha=0.5)
axs[0,0].set_xlabel("Ciudades")
axs[0,0].set_ylabel("Habitantes (miles)")
axs[0,0].set_title("Ciudades más pobladas de California")
for i, bar in enumerate(axs[0, 0].containers[0]):
    height = bar.get_height()
    axs[0,0].text(x=bar.get_x() + bar.get_width() /2, y=height +100000, s=height/1000, ha="center",va="bottom")


axs[0,1].hist(dataset["area_total_km2"], bins=50, alpha=0.5, color="steelblue")
axs[0,1].set_xlabel("Superficie (km2)")
axs[0,1].set_ylabel("Número de ciudades")
axs[0,1].legend(["Hist area"], loc="upper right", title="Hist area")


axs[1,0].scatter(dataset["area_water_km2"], dataset["population_total"], c="g",alpha=0.3)
axs[1,0].set_xlabel("Área de agua en la ciudad (km2)")
axs[1,0].set_ylabel("Población (miles habitantes)")
axs[1,0].set_title("Población por kilómetro cuadrado de agua")


axs[1,1].boxplot(dataset["elevation_m"], vert=False, patch_artist=True, boxprops=dict(facecolor='red',color='black'))
axs[1,1].set_xlabel("Altitud (m)")
axs[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California")


plt.tight_layout()
plt.show()
```
